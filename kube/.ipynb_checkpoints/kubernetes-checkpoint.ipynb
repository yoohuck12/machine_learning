## Kubernetes 구조 이해하기

### Node

Multiple pod 를 가지고 있는 객체라고 생각하면 되고, 3개의 process 가 기본적으로 동작한다.
* Container Runtime 
* Kublet: Container 와 node 사이에서 동작하는 프로그램이며, Master node 의 요청에 의해서 실제로 container 를 실행시킴.
* kube proxy: Request 를 forwarding 해주는 역할을 함.

### Master

k8s 의 모든 노드를 관리하고, 실제로 유저와의 interaction 을 담당하며 다양한 역할 을 수행한다.
총 4가지의 역할을 기본적으로 한다.
* API server: 유저의 요청을 받아서 처리하는 역할
* Scheduler: 노드의 상태를 보고 새로운 job 을 어디에 실행시킬지에 대해서 결정하는 역할
* Controller manager: 노드가 죽었을때 self-healing 을 수행하고, 어떤 노드에 실질적으로 이미지를 분배시키는 역할을 하는 프로세스
* etcd: 다양한 메타 데이터를 저장해두고 있는 저장소

## YAML 의 구조 이해하기

YAML 은 "YAML은 마크업 언어가 아니다 (YAML Ain't Markup Language)” 라는 줄임말인데, '가벼운 마크업 언어' 로써 사용되고 있다고 보면 될 것 같다. YAML 은 기본적으로 KEY, VALUE 로 이루어진 구조를 가지고 있으며, KEY, VALUE 를 재귀적으로 hierachical 하게 사용할 수 있으며, 리스트 와 구조체와 같은 메모리 구조와 더불어 변수까지도 사용할 수 있게 되어 있다.

YAML 문법에서 특히 중요한 것은 들여쓰기로 잘못된 들여쓰기를 하게 되면 하고자 하는 바가 완전히 달라질 수 있어서 조심해야 한다.

YAML 예제 파일.

```YAML
person:
  name: &name yoohyuck
  age: !!float 32
  height: !!str 173.5
  fav_num: 1e+10
  birthday: 1990-11-10 17:00:00
  hobby
    - soccer
    - baseball
  Movies: ['hunger game', 'hobit', 'harry potter']
  friends:
    - name: 'Kim soo young'
      age: 31
  description: >  # Only one line.
    I'm a boy with passion, and I have something to
    have in mind. Everything gonna be okay and I will
    gonna success.
  signature: |  # Keep newline.
    Yoohyuck
    email - yoohuck12@gmail.com
  id: *name
  base: &base
    var1: value1
  foo: 
    <<: *base  # var1: value1
```



## Kubenetes YAML 구조 파악하기

위의 정보를 토대로 우리는 Kubernetes 의 간단한 YAML 파일을 살펴보고, 어떤 구조로 생겼는지 알아보고자 한다.

기본적으로 모든 YAML 파일은 3개의 값을 무조건 가지고 있다. 2개는 사용자에 의해서 설정되고 1개는 자동으로 만들어진다.
* (수동) metadata: 자신에 대한 설명. name + labels
* (수동) specification: 자신이 하는 일에 대한 스펙. kind 별로 어떤 component 에 대해서 세팅할지에 대한 키, 값들이 달라짐
* (자동) status: desired status vs actual status 를 비교해서 실제로 원하는 status 를 달성하고 있는지 확인하기 위해서 사용됨.

![status_image](./deployment_status.jpg)

[Deployment](https://kubernetes.io/ko/docs/concepts/workloads/controllers/deployment/)
```YAML
apiVersion: apps/v1
kind: Deployment
metadata:
  name: nginx-deployment
  labels:
    app: nginx
spec:
  replicas: 3
  selector:
    matchLabels:
      app: nginx
  template:
    metadata:
      labels:
        app: nginx
    spec:
      containers:
      - name: nginx
        image: nginx:1.14.2
        ports:
        - containerPort: 80
```

[Service](https://kubernetes.io/ko/docs/concepts/services-networking/service/)
```YAML
apiVersion: v1
kind: Service
metadata:
  name: my-service
spec:
  selector:
    app: MyApp
  ports:
    - protocol: TCP
      port: 80
      targetPort: 9376
```
